**Q1:-What is hypothesis testing in statistics?**

**Ans:-**
Hypothesis testing is a fundamental method in statistics used to make decisions or inferences about population parameters based on sample data.


**Key Concepts:**

Null Hypothesis (H₀): A default assumption that there is no effect or no difference.

Alternative Hypothesis (H₁ or Ha): The hypothesis that there is an effect or a difference.

Test Statistic: A value computed from the sample data that is used to decide whether to reject H₀.

P-value: The probability of observing the test statistic or something more extreme under H₀.

Significance Level (α): The threshold for rejecting H₀, commonly 0.05.

**Q2:-What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Ans:-**
In statistics, hypothesis testing is used to make decisions about data. It involves two competing statements:

🔹 Null Hypothesis (H₀)
Represents no effect, no difference, or status quo.

It is the hypothesis that the test seeks to disprove.

Example: "The average height of a population is 170 cm."

🔹 Alternative Hypothesis (H₁ or Ha)
Represents a new claim, an effect, or a difference.

It is what you want to prove or support.

Example: "The average height of a population is not 170 cm."

**Q3:-What is the significance level in hypothesis testing, and why is it important?**

**Ans:-**The significance level in hypothesis testing, often denoted by α (alpha), is the probability of rejecting the null hypothesis when it is actually true (i.e., a Type I error). It is a threshold used to determine whether the observed data is statistically significant.

**Why It’s Important:**

It controls the risk of making a false positive conclusion.

Common values: α = 0.05 (5%), α = 0.01 (1%), or α = 0.10 (10%).

A smaller α means you're less likely to reject the null hypothesis, reducing the chance of a Type I error but increasing the chance of a Type II error (false negative).

**Q4:-What does a P-value represent in hypothesis testing?**

**Ans:-**The P-value is the probability of observing a test statistic as extreme as the one computed assuming the null hypothesis is true. It helps determine whether the observed results are statistically significant.

**Interpretation of the P-value:**

Small p-value (≤ α): Strong evidence against the null hypothesis → Reject the null.

Large p-value (> α): Weak evidence against the null → Fail to reject the null.

Typical significance level α = 0.05.

**Q5:-How do you interpret the P-value in hypothesis testing?**

**Ans:-**How to Interpret the P-value in Hypothesis Testing (Theoretical Explanation with Python Context)

**The P-value**

The P-value is the probability of obtaining results at least as extreme as the observed data, assuming the null hypothesis H0 is true.

It answers this question:

"If the null hypothesis were true, how likely is it to see a test statistic as extreme as the one we observed?"

**Steps to Interpret the P-value (in Theory)**

Alternative hypothesis H1: Assumes there is an effect or a difference.

Set a Significance Level α (commonly 0.05)

Compute the P-value using a statistical test (e.g., t-test, z-test, chi-squared test).

Compare P-value with α:

If P-value≤𝛼:

➤ Reject H0

➤ The result is statistically significant

If P-value>𝛼:

➤ Fail to reject H0

➤ The result is not statistically significant

**Important Notes:**

A small P-value means the observed result is unlikely under H0.

A large P-value means the observed result is likely under H0.

P-value ≠ probability that the null hypothesis is true.

P-value ≠ magnitude of the effect.

**Q10:-What is the T-distribution, and when should it be used instead of the normal distribution?**

**Q14:-What is a confidence interval, and how is it used to interpret statistical results?**

**Ans:-**A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like the mean) with a certain level of confidence.

For example, a 95% confidence interval means that if you repeated the sampling many times, about 95% of those intervals would contain the true population mean.

It gives not just a single estimate (like a sample mean), but a range which reflects the uncertainty due to sampling.

**How is it used to interpret statistical results**
Instead of just giving a point estimate, the CI gives a range that tells you how precise that estimate is.

If the CI for a mean difference does not include zero, it suggests there is a statistically significant difference at the chosen confidence level.

The width of the CI indicates the level of uncertainty — wider intervals mean more uncertainty.

Interpretation:
The true population mean is estimated to lie between the lower and upper bounds with 95% confidence.

If you were testing if the population mean equals some value (e.g., 2.5), and that value is not inside the confidence interval, you would reject the null hypothesis at the 5% significance level.

**Q15:-What is the margin of error, and how does it affect the confidence interval?**

The margin of error (ME) is the amount you add and subtract from the sample estimate (like the sample mean) to create the confidence interval.

A larger margin of error results in a wider confidence interval, indicating more uncertainty.

Sample size increases (because standard error decreases).

Confidence level decreases (lower critical value).

```python
#Q1:- Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm

# Function to perform one-sample Z-test
def z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate the Z-score
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate the p-value
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z)))
    elif alternative == 'less':
        p_value = norm.cdf(z)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    # Interpret the result
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# Example usage
sample = [100, 102, 98, 97, 101, 99, 96, 105, 103, 104]  # sample data
population_mean = 100                                   # known population mean
population_std = 5                                       # known population std dev

z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided')
```

```python
#Q3:- Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample_data (list or array): Sample observations
        population_mean (float): Hypothesized population mean
        population_std (float): Known population standard deviation
        alpha (float): Significance level (default 0.05)

    Returns:
        z_score (float): Calculated Z-score
        p_value (float): P-value for the test
        conclusion (str): Result interpretation
    """
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Z-test statistic
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed test p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Conclusion
    if p_value < alpha:
        conclusion = "Reject the null hypothesis (significant difference)."
    else:
        conclusion = "Fail to reject the null hypothesis (no significant difference)."

    return z_score, p_value, conclusion

# Example usage
sample = [51, 49, 50, 52, 48, 47, 53, 50, 49, 52]
population_mean = 50
population_std = 5

z, p, result = one_sample_z_test(sample, population_mean, population_std)
print(f"Z-score: {z:.4f}")
print(f"P-value: {p:.4f}")
print("Conclusion:", result)
```

```python
#Q4:-Perform a two-tailed Z-test using Python and visualize the decision region on a plot?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given parameters
mu = 100        # Population mean
sigma = 15      # Population std dev
x_bar = 108     # Sample mean
n = 36          # Sample size
alpha = 0.05    # Significance level

# Z-test statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))
print(f"Z-score: {z:.2f}")

# Two-tailed critical z-values
z_critical = stats.norm.ppf(1 - alpha / 2)
print(f"Critical z-value for two-tailed test: ±{z_critical:.2f}")

# P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}")

# Plotting the decision region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')

# Fill critical regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (α/2)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4)

# Plot observed Z-score
plt.axvline(z, color='blue', linestyle='--', label=f'Observed Z = {z:.2f}')

# Formatting
plt.title('Two-tailed Z-Test with Rejection Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q5:-Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, tail='two-tailed'):
    """
    Visualize Type I and Type II errors for a hypothesis test.

    Parameters:
    - mu0: Mean under H0
    - mu1: Mean under H1
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    - tail: 'left-tailed', 'right-tailed', or 'two-tailed'
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Compute critical values
    if tail == 'two-tailed':
        z_critical = stats.norm.ppf(1 - alpha / 2)
        lower_crit = mu0 - z_critical * se
        upper_crit = mu0 + z_critical * se
    elif tail == 'left-tailed':
        z_critical = stats.norm.ppf(alpha)
        lower_crit = mu0 + z_critical * se
        upper_crit = np.inf
    elif tail == 'right-tailed':
        z_critical = stats.norm.ppf(1 - alpha)
        lower_crit = -np.inf
        upper_crit = mu0 + z_critical * se
    else:
        raise ValueError("tail must be 'left-tailed', 'right-tailed', or 'two-tailed'")

    # Range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions
    y0 = stats.norm.pdf(x, mu0, se)
    y1 = stats.norm.pdf(x, mu1, se)

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(x, y0, label='H0: μ = {:.2f}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H1: μ = {:.2f}'.format(mu1), color='green')

    # Type I Error (α): Reject H0 when it's actually true
    plt.fill_between(x, y0, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.3, label='Type I Error (α)')

    # Type II Error (β): Fail to reject H0 when H1 is true
    if tail == 'two-tailed':
        plt.fill_between(x, y1, where=(x >= lower_crit) & (x <= upper_crit), color='orange', alpha=0.3, label='Type II Error (β)')
    elif tail == 'left-tailed':
        plt.fill_between(x, y1, where=(x >= lower_crit), color='orange', alpha=0.3, label='Type II Error (β)')
    elif tail == 'right-tailed':
        plt.fill_between(x, y1, where=(x <= upper_crit), color='orange', alpha=0.3, label='Type II Error (β)')

    # Labels and legend
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    if np.isfinite(upper_crit):
        plt.axvline(upper_crit, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate and print beta (Type II error)
    if tail == 'two-tailed':
        beta = stats.norm.cdf((upper_crit - mu1)/se) - stats.norm.cdf((lower_crit - mu1)/se)
    elif tail == 'left-tailed':
        beta = 1 - stats.norm.cdf((lower_crit - mu1)/se)
    elif tail == 'right-tailed':
        beta = stats.norm.cdf((upper_crit - mu1)/se)

    power = 1 - beta
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Statistical Power (1 - β): {power:.4f}")
```

```python
#Q6:-Write a Python program to perform an independent T-test and interpret the results?
import numpy as np
from scipy import stats

def perform_independent_ttest(group1, group2, alpha=0.05):
    """
    Perform an independent t-test between two groups and interpret the result.

    Parameters:
    - group1, group2: Arrays or lists of sample data
    - alpha: Significance level (default is 0.05)
    """

    # Convert to NumPy arrays
    group1 = np.array(group1)
    group2 = np.array(group2)

    # Perform independent two-sample t-test
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Welch's t-test by default

    # Output results
    print("=== Independent T-Test Results ===")
    print(f"Group 1 Mean: {np.mean(group1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(f"\n Reject the null hypothesis (p < {alpha}).")
        print("There is a statistically significant difference between the two groups.")
    else:
        print(f"\n Fail to reject the null hypothesis (p ≥ {alpha}).")
        print("There is no statistically significant difference between the two groups.")

# Example data (replace with your own samples)
group1 = [23, 21, 19, 24, 30, 18, 20]
group2 = [31, 35, 29, 27, 33, 30, 32]

perform_independent_ttest(group1, group2, alpha=0.05)
```

```python
#Q7:-Perform a paired sample T-test using Python and visualize the comparison results?
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def perform_paired_ttest(before, after, alpha=0.05):
    """
    Perform a paired sample t-test and visualize the results.

    Parameters:
    - before: Measurements before treatment/intervention
    - after: Measurements after treatment/intervention
    - alpha: Significance level (default = 0.05)
    """

    # Convert to NumPy arrays
    before = np.array(before)
    after = np.array(after)

    # Check that the lengths match
    if len(before) != len(after):
        raise ValueError("Both samples must have the same number of observations.")

    # Perform paired t-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Print results
    print("=== Paired Sample T-Test Results ===")
    print(f"Mean (Before): {np.mean(before):.2f}")
    print(f"Mean (After): {np.mean(after):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"\n Reject the null hypothesis (p < {alpha})")
        print("There is a significant difference between the paired samples.")
    else:
        print(f"\n Fail to reject the null hypothesis (p ≥ {alpha})")
        print("There is no significant difference between the paired samples.")

    # Visualization
    indices = np.arange(len(before))
    width = 0.35

    plt.figure(figsize=(10, 6))
    plt.bar(indices - width/2, before, width, label='Before', alpha=0.7)
    plt.bar(indices + width/2, after, width, label='After', alpha=0.7)

    for i in range(len(before)):
        plt.plot([i - width/2, i + width/2], [before[i], after[i]], color='gray', linestyle='--', linewidth=0.8)

    plt.xlabel('Subject')
    plt.ylabel('Measurement')
    plt.title('Before vs After: Paired Sample Comparison')
    plt.xticks(indices)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Example data (replace with real measurements)
before = [85, 90, 88, 75, 95, 78, 82, 89, 91, 84]
after =  [88, 94, 89, 78, 97, 80, 86, 91, 92, 86]

perform_paired_ttest(before, after, alpha=0.05)
```

```python
#Q8:-Simulate data and perform both Z-test and T-test, then compare the results using Python?
import numpy as np
import scipy.stats as stats

def simulate_data_and_compare_tests(mu=100, sigma=15, sample_size=30, alpha=0.05):
    np.random.seed(0)  # For reproducibility

    # Simulate data from a normal distribution
    sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)

    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # unbiased sample std

    # === Z-Test ===
    # Known population standard deviation
    se_known = sigma / np.sqrt(sample_size)
    z_stat = (sample_mean - mu) / se_known
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # === T-Test ===
    # Unknown population standard deviation
    t_stat, p_value_t = stats.ttest_1samp(sample, popmean=mu)

    # === Results ===
    print("=== Simulated Data Summary ===")
    print(f"Sample Size: {sample_size}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}\n")

    print("=== Z-Test (σ known) ===")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value_z:.4f}")
    print(f"{'Reject' if p_value_z < alpha else 'Fail to reject'} the null hypothesis at α = {alpha}\n")

    print("=== T-Test (σ unknown) ===")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value_t:.4f}")
    print(f"{'Reject' if p_value_t < alpha else 'Fail to reject'} the null hypothesis at α = {alpha}\n")

    print("=== Comparison ===")
    print(f"Difference in P-values: {abs(p_value_z - p_value_t):.4f}")
    print("Note: With small sample sizes, T-test is more conservative.\n")

# Run it
simulate_data_and_compare_tests()
```

```python
#Q9:-Write a Python function to calculate the confidence interval for a sample mean and explain its significance?
import numpy as np
import scipy.stats as stats

def confidence_interval_mean(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample: array-like, sample data
    - confidence: float, confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): Tuple of confidence interval
    """

    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    sem = stats.sem(sample)  # Standard error of the mean

    # t-critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
data = [50, 52, 49, 51, 53, 54, 48, 47, 52, 50]
ci_lower, ci_upper = confidence_interval_mean(data, confidence=0.95)

print(f"95% Confidence Interval for the mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q10:-Write a Python program to calculate the margin of error for a given confidence level using sample data?

import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
    - sample: list or array-like, sample data
    - confidence: float, confidence level (default is 0.95)

    Returns:
    - margin_of_error: float
    """
    sample = np.array(sample)
    n = len(sample)
    if n < 2:
        raise ValueError("Sample must contain at least two values.")

    sample_std = np.std(sample, ddof=1)
    sem = sample_std / np.sqrt(n)  # Standard Error of the Mean

    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * sem

    return margin_of_error

# Example usage
data = [50, 52, 49, 51, 53, 54, 48, 47, 52, 50]
confidence_level = 0.95

moe = calculate_margin_of_error(data, confidence=confidence_level)
print(f"Margin of Error at {int(confidence_level*100)}% confidence: ±{moe:.2f}")
```

```python
#Q11:-Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?
def bayesian_inference(prior_disease, sensitivity, specificity):
    """
    Uses Bayes' Theorem to calculate the posterior probability of having a disease
    given a positive test result.

    Parameters:
    - prior_disease: float, prior probability of having the disease (P(D))
    - sensitivity: float, true positive rate (P(+|D))
    - specificity: float, true negative rate (P(-|¬D))

    Returns:
    - posterior: float, P(D|+)
    """

    # Prior probabilities
    p_disease = prior_disease
    p_no_disease = 1 - p_disease

    # Likelihoods
    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity  # false positive rate

    # Total probability of a positive test (denominator)
    p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Bayes' Theorem: Posterior
    posterior = (p_pos_given_disease * p_disease) / p_positive

    return posterior

# Example
prior = 0.01         # 1% of people have the disease
sensitivity = 0.99   # test correctly identifies 99% of true cases
specificity = 0.99   # test correctly excludes 99% of healthy people

posterior = bayesian_inference(prior, sensitivity, specificity)

print(f"Probability of having the disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

```python
#Q13:-Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data?
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test given observed data.

    Parameters:
    - observed: 2D numpy array or list of lists (contingency table)

    Returns:
    - expected: 2D numpy array of expected frequencies
    """

    observed = np.array(observed)
    row_sums = observed.sum(axis=1).reshape(-1, 1)  # sum of each row (as column vector)
    col_sums = observed.sum(axis=0).reshape(1, -1)  # sum of each column (as row vector)
    total = observed.sum()

    expected = (row_sums @ col_sums) / total  # matrix multiplication for expected counts

    return expected

# Example observed data (contingency table)
observed_data = [
    [30, 10],
    [20, 40]
]

expected_freq = calculate_expected_frequencies(observed_data)
print("Observed Frequencies:\n", np.array(observed_data))
print("\nExpected Frequencies:\n", expected_freq)
```

```python
#Q16:-Implement an F-test using Python to compare the variances of two random samples?
import numpy as np
from scipy.stats import f

def f_test_variances(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1, sample2: array-like, the two samples
    - alpha: significance level (default 0.05)

    Returns:
    - f_stat: F-statistic value
    - p_value: two-tailed p-value for the test
    - conclusion: interpretation string
    """

    # Sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Sample sizes
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate F-statistic (larger variance / smaller variance)
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1  # degrees of freedom numerator and denominator
    else:
        f_stat = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Two-tailed p-value
    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    # Conclusion
    if p_value < alpha:
        conclusion = "Reject null hypothesis: variances are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: no significant difference in variances."

    return f_stat, p_value, conclusion

# Example usage
np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=1.0, size=30)  # variance ~1
sample2 = np.random.normal(loc=0, scale=1.5, size=30)  # variance ~2.25

f_stat, p_value, conclusion = f_test_variances(sample1, sample2)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
```

```python
#Q17:-Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results?
import numpy as np
from scipy.stats import f_oneway

def perform_anova(*groups, alpha=0.05):
    """
    Perform one-way ANOVA on multiple groups.

    Parameters:
    - groups: variable number of array-like groups
    - alpha: significance level (default 0.05)

    Returns:
    - f_stat: F-statistic
    - p_value: p-value of the test
    - conclusion: interpretation string
    """
    f_stat, p_value = f_oneway(*groups)

    if p_value < alpha:
        conclusion = "Reject null hypothesis: At least one group mean is significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: No significant difference between group means."

    return f_stat, p_value, conclusion

# Example usage:
np.random.seed(0)

group1 = np.random.normal(20, 5, size=30)
group2 = np.random.normal(22, 5, size=30)
group3 = np.random.normal(19, 5, size=30)

f_stat, p_value, conclusion = perform_anova(group1, group2, group3)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
```

```python
#Q19:-Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA?
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check ANOVA assumptions:
    - Normality of each group (Shapiro-Wilk test)
    - Homogeneity of variances (Levene’s test)
    - Independence: cannot be tested statistically here, reminder only

    Parameters:
    - groups: variable number of array-like groups
    - alpha: significance level (default 0.05)

    Returns:
    - results: dict with keys 'normality', 'homogeneity', 'independence_note'
    """
    results = {}

    # 1. Normality test for each group
    normality_results = {}
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        normality_results[f'Group {i}'] = {
            'W-statistic': stat,
            'p-value': p,
            'normal': p > alpha
        }
    results['normality'] = normality_results

    # 2. Homogeneity of variances (Levene’s test)
    stat, p = levene(*groups)
    results['homogeneity'] = {
        'Levene_statistic': stat,
        'p-value': p,
        'equal_variance': p > alpha
    }

    # 3. Independence reminder (cannot test from data)
    results['independence_note'] = (
        "Independence of observations should be ensured by study design; "
        "it cannot be tested statistically here."
    )

    return results

# Example usage:
import numpy as np

np.random.seed(42)
g1 = np.random.normal(20, 5, 30)
g2 = np.random.normal(22, 5, 30)
g3 = np.random.normal(19, 5, 30)

assumption_results = check_anova_assumptions(g1, g2, g3)

for group, res in assumption_results['normality'].items():
    print(f"{group} Normality: W={res['W-statistic']:.4f}, p={res['p-value']:.4f} -> Normal? {res['normal']}")

levene_res = assumption_results['homogeneity']
print(f"\nLevene's test for equal variance: Stat={levene_res['Levene_statistic']:.4f}, p={levene_res['p-value']:.4f} -> Equal variance? {levene_res['equal_variance']}")

print(f"\nIndependence note: {assumption_results['independence_note']}")
```

```python
#Q20:-Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results?
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from statsmodels.graphics.factorplots import interaction_plot

# Simulate data
np.random.seed(0)
levels_A = ['Low', 'High']
levels_B = ['Type1', 'Type2']

data = []

for a in levels_A:
    for b in levels_B:
        # Mean depends on combination to simulate interaction effect
        if a == 'Low' and b == 'Type1':
            mean = 10
        elif a == 'Low' and b == 'Type2':
            mean = 12
        elif a == 'High' and b == 'Type1':
            mean = 15
        else:  # High & Type2
            mean = 18

        # Generate 10 samples per group
        samples = np.random.normal(loc=mean, scale=2, size=10)
        for value in samples:
            data.append([a, b, value])

# Create DataFrame
df = pd.DataFrame(data, columns=['FactorA', 'FactorB', 'Value'])

# Fit two-way ANOVA model
model = ols('Value ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-way ANOVA results:")
print(anova_table)

# Interaction plot
plt.figure(figsize=(8,6))
interaction_plot(df['FactorA'], df['FactorB'], df['Value'],
                 colors=['red', 'blue'], markers=['o', 's'], ms=8)
plt.title('Interaction Plot: FactorA and FactorB')
plt.xlabel('FactorA')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```

```python
#Q21:-Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(dfn, dfd):
    """
    Plot the F-distribution PDF for given degrees of freedom.

    Parameters:
    - dfn: degrees of freedom numerator
    - dfd: degrees of freedom denominator
    """
    x = np.linspace(0, 5, 500)
    y = f.pdf(x, dfn, dfd)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'b-', lw=2, label=f'F-distribution\n(df1={dfn}, df2={dfd})')
    plt.fill_between(x, y, alpha=0.3)
    plt.title('F-distribution Probability Density Function')
    plt.xlabel('F-value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    # Degrees of freedom for numerator and denominator
    dfn = 5  # e.g., number of groups - 1
    dfd = 20  # e.g., total observations - number of groups

    plot_f_distribution(dfn, dfd)

    print(f"""
    The F-distribution arises in hypothesis testing when comparing variances,
    such as in ANOVA tests. It is the distribution of the ratio of two scaled chi-square variables.

    In hypothesis testing, the null hypothesis often states that group variances or means are equal.
    The F-statistic calculated from data is compared against critical values from this distribution.

    If the F-statistic exceeds the critical value at a chosen significance level (e.g., 0.05),
    we reject the null hypothesis, indicating at least one group differs significantly.
    """)

if __name__ == "__main__":
    main()
```

```python
#Q24:-Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results?
import numpy as np
from scipy.stats import chi2

def variance_hypothesis_test(sample, hypothesized_variance, alpha=0.05):
    n = len(sample)
    sample_variance = np.var(sample, ddof=1)

    # Test statistic
    chi_square_stat = (n - 1) * sample_variance / hypothesized_variance

    # Degrees of freedom
    df = n - 1

    # Two-tailed critical values
    chi2_lower = chi2.ppf(alpha / 2, df)
    chi2_upper = chi2.ppf(1 - alpha / 2, df)

    # p-value calculation for two-tailed test
    if chi_square_stat < df:
        p_value = 2 * chi2.cdf(chi_square_stat, df)
    else:
        p_value = 2 * (1 - chi2.cdf(chi_square_stat, df))
    p_value = min(p_value, 1)  # p-value cannot be > 1

    # Interpretation
    if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
        conclusion = "Reject null hypothesis: population variance is significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: no significant difference in variance."

    return {
        'sample_variance': sample_variance,
        'chi_square_statistic': chi_square_stat,
        'chi2_lower_critical': chi2_lower,
        'chi2_upper_critical': chi2_upper,
        'p_value': p_value,
        'conclusion': conclusion
    }

# Example usage
np.random.seed(123)
sample_data = np.random.normal(loc=0, scale=5, size=30)  # true variance = 25
hypothesized_var = 20  # test against this variance

result = variance_hypothesis_test(sample_data, hypothesized_var)
print(f"Sample variance: {result['sample_variance']:.4f}")
print(f"Chi-square statistic: {result['chi_square_statistic']:.4f}")
print(f"Chi-square critical values: [{result['chi2_lower_critical']:.4f}, {result['chi2_upper_critical']:.4f}]")
print(f"P-value: {result['p_value']:.4f}")
print(f"Conclusion: {result['conclusion']}")
```

```python
#Q25:-Write a Python script to perform a Z-test for comparing proportions between two datasets or groups?
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_ztest(count1, nobs1, count2, nobs2, alpha=0.05):
    """
    Perform a two-proportion Z-test.

    Parameters:
    - count1: number of successes in group 1
    - nobs1: total observations in group 1
    - count2: number of successes in group 2
    - nobs2: total observations in group 2
    - alpha: significance level

    Returns:
    - z_stat: z test statistic
    - p_value: p-value of the test
    - conclusion: interpretation string
    """
    counts = np.array([count1, count2])
    nobs = np.array([nobs1, nobs2])

    z_stat, p_value = proportions_ztest(counts, nobs)

    if p_value < alpha:
        conclusion = "Reject null hypothesis: proportions are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: insufficient evidence of difference."

    return z_stat, p_value, conclusion

# Example usage:
# Group 1: 45 successes out of 100 trials
# Group 2: 30 successes out of 100 trials
z_stat, p_value, conclusion = two_proportion_ztest(45, 100, 30, 100)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
```

```python
#Q26:-Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create or input two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)  # Dataset 1
data2 = np.random.normal(loc=50, scale=20, size=100)  # Dataset 2 (higher variance)

# Step 2: Calculate variances and F-statistic
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# F = larger variance / smaller variance
F = var2 / var1 if var2 > var1 else var1 / var2

# Degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 3: Calculate p-value (two-tailed)
p_value = 2 * min(stats.f.cdf(F, df1, df2), 1 - stats.f.cdf(F, df1, df2))

# Step 4: Print results
print("Variance of data1:", round(var1, 2))
print("Variance of data2:", round(var2, 2))
print("F-statistic:", round(F, 4))
print("Degrees of freedom:", df1, "and", df2)
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: the variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: no significant difference in variances.")

# Step 5: Visualization
plt.figure(figsize=(10, 5))
sns.histplot(data1, kde=True, color='skyblue', label='Data1 (Low Var)', stat='density')
sns.histplot(data2, kde=True, color='salmon', label='Data2 (High Var)', stat='density')
plt.title('Distribution Comparison')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q27:-Perform a Chi-square test for goodness of fit with simulated data and analyze the results?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (e.g., dice rolls)
np.random.seed(0)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120, replace=True)

# Count occurrences for each outcome
observed_freq = np.bincount(observed)[1:]  # Ignore zero index

# Step 2: Define expected frequencies (uniform distribution)
expected_freq = np.full_like(observed_freq, fill_value=len(observed) / 6)

# Step 3: Perform chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

# Step 4: Print results
print("Observed frequencies:", observed_freq)
print("Expected frequencies:", expected_freq)
print("Chi-square Statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: observed distribution differs from expected.")
else:
    print("Fail to reject the null hypothesis: no significant difference from expected.")

# Step 5: Visualization
categories = np.arange(1, 7)
bar_width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(categories - bar_width/2, observed_freq, width=bar_width, label='Observed', color='skyblue')
plt.bar(categories + bar_width/2, expected_freq, width=bar_width, label='Expected', color='salmon')
plt.xlabel('Category (Dice Face)')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness-of-Fit: Observed vs Expected')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(categories)
plt.tight_layout()
plt.show()
```
